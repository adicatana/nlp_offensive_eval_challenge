```
def parse_in(tokenized_corpus, word2idx, labels, max_len):
    # we index our sentences
    vectorized_sentences = [[word2idx[token] for token in sentence if token in word2idx] for sentence in tokenized_corpus]
  
  
    # we create a tensor of a fixed size filled with zeroes for padding
    sentences_tensor = Variable(torch.zeros((len(vectorized_sentences), max_len))).long()
    sentences_lengths = [len(sentence) for sentence in vectorized_sentences]

    # we fill it with our vectorized sentences 
    for idx, (sentence, sentence_len) in enumerate(zip(vectorized_sentences, sentences_lengths)):
        sentences_tensor[idx, :sentence_len] = torch.LongTensor(sentence)
            
    labels_tensor = torch.FloatTensor(labels)

    return sentences_tensor, labels_tensor
```

```
def train_model_custom(model, optimizer, loss_fn, feature_train, target_train, feature_valid, target_valid, feature_test, target_test):
#     target_train = target_train.type(torch.LongTensor)
#     target_valid = target_valid.type(torch.LongTensor)
    for epoch in range(1, epochs+1):
        model.train()
        optimizer.zero_grad()
        predictions = model(feature_train)
#         print(predictions)
#         print(predictions.shape)
#         print(target_train)
#         print(target_train.shape)
        loss = loss_fn(predictions, target_train.type(torch.LongTensor))
        acc = accuracy_task_c(predictions, target_train.type(torch.LongTensor))
        loss.backward()
        optimizer.step()

        epoch_loss = loss.item()
        epoch_acc = acc
  
        model.eval()
  
        with torch.no_grad():

            predictions_valid = model(feature_valid)
            loss = loss_fn(predictions_valid, target_valid.type(torch.LongTensor))
            acc = accuracy_task_c(predictions_valid, target_valid.type(torch.LongTensor))
            valid_loss = loss.item()
            valid_acc = acc

        print(f'| Epoch: {epoch:02} | Train Loss: {epoch_loss:.3f} | Train Acc: {epoch_acc*100:.2f}% | Val. Loss: {valid_loss:.3f} | Val. Acc: {valid_acc*100:.2f}% |')
    
    model.eval()

    with torch.no_grad():
 
        predictions = model(feature_test)
        loss = loss_fn(predictions, torch.tensor(target_test).type(torch.LongTensor))
        acc = accuracy_task_c(predictions, torch.tensor(target_test).type(torch.LongTensor))
        print(f'| Test Loss: {loss:.3f} | Test Acc: {acc*100:.2f}%')
        f_measure_task_c(predictions, target_test)

    print(np.count_nonzero(target_test))
```
